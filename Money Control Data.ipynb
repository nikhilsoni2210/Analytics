```python
#file = "Details.csv"
#f = open(file, "w")
Headers = "Date,Recommendation"
#f.write(Headers)
list_of_recommendation = []
for page in range(100,200):
    url = "https://www.moneycontrol.com/broker-research/markets/cash-"+str(page)+".html".format(page)
    html = urlopen(url)
    soup = BeautifulSoup(html,"html.parser")
    Title = soup.find_all("div", {"class":"Ohidden"})
    
    for i in Title:
        try:
            date = i.find("p", {"class":"op_gl12"}).get_text()
            recommendation = i.find("p", {"class":"MT5"}).get_text()
            list_of_recommendation.append((date,recommendation))
            
        except:
            pass
```

```python
list_100_200 = list_of_recommendation 
```

```python
for i in range(len(list_100_200)):
    a = list_100_200[i].lower()
    if ":" in a:
        a = a.replace(":",";")
    if "accumulate" in a:
        a = a.replace("accumulate","Buy")
    if "Pick" in a:
        a = a.replace("pick","Buy")
    list_100_200[i] = a
    
```
